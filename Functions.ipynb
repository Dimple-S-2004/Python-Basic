An iterator must have two methods:

6. Explain the concept of generator in Python and how they are defined.

-->A generator is a special type of iterator in Python.

Instead of storing all values in memory (like a list), it generates values one by one as you loop through it.

Generators are more memory-efficient because they don’t create the entire sequence at once.

Example: def my_gen():
             yield 1
             yield 2
             yield 3

   for i in my_gen():
        print(i)

7. What are the advantages of using generators over regular functions?

--> Advantages of using generators:
1. Memory Efficiency:
A generator produces one item at a time, instead of creating the entire list in memory.
2. Lazy Evaluation:
Values are generated only when needed, not in advance.
Saves computation when you don’t actually need all results.
3. Convenient State Saving:
With yield, a generator remembers where it left off automatically.
No need for manual state management like in normal functions with loops.
4. Readable and Simple Code:
Generators make code cleaner than manually writing iterator classes with __iter__ and __next__.
Generators can represent infinite sequences (like natural numbers).

9. Expalin the purpose and usage of the map() functions in Python.

--->The map() function applies a given function to each item in an iterable (like list, tuple, etc.) and returns a map object (which is an iterator).

Syntax:
map(function, iterable)

Purpose of map():

To avoid writing explicit loops when applying the same operation to all elements of a sequence.

Makes the code shorter and more readable.

Works efficiently with large data since it returns an iterator (lazy evaluation).

Example:nums = [10, 20, 30]
result = map(lambda x: x + 5, nums)
print(list(result))   # [15, 25, 35]

10. What is the difference between map(), reduce(), an filter() function in python?

--->1. map()

Purpose: Apply a function to each element of an iterable.

Result: Returns a new iterator with transformed values.

Example:
nums = [1, 2, 3, 4]
squares = list(map(lambda x: x*x, nums))
print(squares)   # [1, 4, 9, 16]

2. filter()

Purpose: Filter out elements of an iterable based on a condition (function that returns True/False).

Result: Returns only elements that satisfy the condition.

Example:
nums = [10, 15, 20, 25]
evens = list(filter(lambda x: x % 2 == 0, nums))
print(evens)   # [10, 20]

3. reduce()

Purpose: Reduce an iterable to a single value by applying a function cumulatively.

Result: One final value (sum, product, etc.).

reduce() is available in functools module.

 Example:
from functools import reduce

nums = [1, 2, 3, 4]
product = reduce(lambda x, y: x * y, nums)
print(product)   # 24  (1*2*3*4)

```python
#1.Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.
def sum_of_even(numbers):
    return sum(num for num in numbers if num % 2 == 0)

nums = [1, 2, 3, 4, 5, 6]
print(sum_of_even(nums))
```

```python
# 4.Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# Print prime numbers from 1 to 200
primes = [n for n in range(1, 201) if is_prime(n)]
print(primes)
```

```python
# 5.Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
def fibonacci_gen(n):
  a, b = 0, 1
  for _ in range(n):
    yield a
    a, b = b, a + b

#result
for num in fibonacci_gen(7):
    print(num)
```

```python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def power_of_two(n):
  for i in range(n + 1):
    yield 2 ** i

#result
for power in power_of_two(8):
    print(power)
```

```python
#7. Implement a generator function that reads a file line by line and yields each line as a string.
def read_file_line_by_line(filepath):
  with open(filepath, 'r') as file:
    for line in file:
      yield line.strip() #strip removes \n at the end

#usage
for line in read_file_line_by_line('/content/example.txt'):
  print(line)
```

```python
# 10. Create a Python program that uses `filter()` to remove all the vowels from a given string.
def remove_vowels(s):
    vowels = "aeiouAEIOU"
    result = filter(lambda char: char not in vowels, s)
    return ''.join(result)

# Example usage
input_string = "Python is a Programming language"
output_string = remove_vowels(input_string)
print(output_string)
```
