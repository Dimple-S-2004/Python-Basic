# Anomaly Detection & Time Series

1. What is Anomaly Detection? Explain its types (point, contextual, and
collective anomalies) with examples.

-->Anomaly Detection is a data analysis technique used to identify data points, patterns, or observations that deviate significantly from normal behavior in a dataset. These unusual observations are called anomalies, outliers, or novelties.
Anomaly detection is widely used in areas where abnormal behavior may indicate errors, fraud, failures, or rare events.

**Why Anomaly Detection is Important:**

- Detects fraud and security breaches

- Identifies system faults and failures

- Improves data quality

- Helps in early warning systems

**Types of Anomalies:** Anomalies are generally classified into three main types:

**Point Anomalies:** A point anomaly occurs when a single data instance is significantly different from the rest of the data.

**Key Characteristics:**

- Individual abnormal data point

- Easy to detect

- Most common type of anomaly

**Example**:

- A credit card transaction of ₹5,00,000 when the customer usually spends ₹5,000

- A temperature sensor suddenly showing 100°C when normal range is 20–30°C

**Real-World Applications:**

- Credit card fraud detection

- Sensor fault detection

**3.2 Contextual Anomalies:** A contextual anomaly occurs when a data point is normal in one context but abnormal in another.
The anomaly is detected based on contextual attributes such as time, location, or environment.

**Key Characteristics:**

- Depends on context

- Requires contextual information

- Common in time-series data

**Example:**

- Temperature of 30°C is normal in summer but abnormal in winter

- High network traffic during business hours is normal, but abnormal at midnight

**Real-World Applications:**

- Weather monitoring

- Network intrusion detection

- Seasonal sales analysis

**3.3 Collective Anomalies:** A collective anomaly occurs when a group of data points together is anomalous, even if individual points appear normal.

**Key Characteristics:**

- Pattern-based anomaly

- Individual points may look normal

- Detected only when viewed as a sequence or group

**Example:**

- A series of small bank withdrawals in a short time indicating fraud

- Continuous low heart rate readings indicating a medical issue

**Real-World Applications**

- Fraud detection

- Health monitoring

- Log and event sequence analysis



**Conclusion:** Anomaly detection is a crucial technique for identifying abnormal behavior in data. Point anomalies focus on individual outliers, contextual anomalies depend on situational context, and collective anomalies involve unusual patterns formed by a group of observations. Understanding these types helps in selecting appropriate detection techniques and improves decision-making in real-world applications such as fraud detection, security, and system monitoring.

2.  Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of
their approach and suitable use cases.

-->  Isolation Forest, DBSCAN, and Local Outlier Factor are popular techniques used for anomaly detection, but they differ in their core approach, assumptions about data, and best-fit use cases. Understanding these differences helps in selecting the right method for a given real-world problem.

**Comparison:**

| Feature | Isolation Forest | DBSCAN | Local Outlier Factor (LOF) |
|--------|------------------|--------|----------------------------|
| Core Idea | Isolates anomalies using random splits | Detects low-density points as noise | Compares local density to neighbors |
| Method Type | Tree-based ensemble | Density-based clustering | Density-based anomaly detection |
| Distance Measure Required | No | Yes | Yes |
| Scalability | Very high | Medium | Low to medium |
| High-Dimensional Data | Performs well | Performs poorly | Struggles |
| Parameter Sensitivity | Low | High (eps, min_samples) | Medium (n_neighbors) |
| Detects Local Anomalies | Limited | Limited | Very good |
| Handles Large Datasets | Excellent | Moderate | Poor |
| Output | Anomaly score | Clusters + noise points | Anomaly score |
| Typical Use Cases | Fraud detection, intrusion detection | Spatial data, geographic anomalies | Local behavior deviations |



**Conclusion** :
Isolation Forest, DBSCAN, and Local Outlier Factor differ significantly in their detection strategies. Isolation Forest isolates anomalies using random partitioning and excels in large, high-dimensional datasets. DBSCAN identifies anomalies as noise in low-density regions and is best suited for spatial or clustered data. Local Outlier Factor focuses on local density differences, making it ideal for detecting subtle, context-specific anomalies. Choosing the right method depends on data size, structure, dimensionality, and the nature of anomalies.

3.  What are the key components of a Time Series? Explain each with one
example.

--> A time series is a sequence of observations recorded over time at regular intervals (daily, monthly, yearly, etc.). Time series analysis aims to understand patterns in data over time and use them for forecasting and decision-making.

**A time series is typically composed of four key components.:**

1. Trend (T) : The trend component represents the long-term movement or direction in the data over a long period of time. It shows whether the data is increasing, decreasing, or stable.

**Example:**
- A steady increase in online shopping sales over several years due to growth in e-commerce
- A long-term rise in a company’s annual revenue

2. Seasonality (S) : Seasonality refers to repeating patterns that occur at fixed and known intervals (daily, weekly, monthly, yearly).

**Example:**
- Higher ice cream sales in summer every year

- Increased electricity consumption during evenings

3. Cyclical Component (C) : The cyclical component represents long-term fluctuations caused by economic or business cycles. Unlike seasonality, cycles do not have a fixed period.

**Example:**

- Rise and fall of employment levels during economic booms and recessions

- Changes in stock market indices over multiple years

4. Irregular / Random Component (I) : The irregular (random) component includes unpredictable and sudden variations caused by unexpected events.

**Example:**

- Sudden drop in airline bookings due to a pandemic

- Sales spike during an unexpected flash sale

**Summary Table: Components of a Time Series**
| Component | Description | Example |
|----------|-------------|---------|
| Trend | Long-term upward or downward movement | Increasing annual sales |
| Seasonality | Regular repeating pattern | Higher sales in festive season |
| Cyclical | Long-term fluctuations without fixed period | Economic boom and recession |
| Irregular | Random, unpredictable variations | Sales drop due to natural disaster |


**Conclusion**: A time series consists of trend, seasonality, cyclical, and irregular components. Understanding these components helps in selecting appropriate forecasting models and improves accuracy in time-dependent predictions.

4. Define Stationary in time series. How can you test and transform a
non-stationary series into a stationary one?

--> A time series is said to be stationary if its statistical properties remain constant over time.
This means that:

- The mean of the series is constant

- The variance of the series is constant

- The covariance between two time points depends only on the lag, not on time

Stationarity is important because most time-series models (e.g., ARMA, ARIMA) assume the data is stationary.

**Types of Stationarity:**

(a) Strict Stationarity

- The joint distribution of the series does not change over time

- Rarely used in practice

(b) Weak (Second-Order) Stationarity

- Mean is constant

- Variance is constant

- Autocovariance depends only on lag

Weak stationarity is most commonly used in real-world modeling.

**3. How to Test Stationarity:**
**3.1 Visual Inspection**

Time series plot → check for trend or seasonality

Rolling mean and variance → should remain constant

**3.2 Statistical Tests:**

(a) Augmented Dickey-Fuller (ADF) Test

Null hypothesis (H₀): Series is non-stationary

If p-value < 0.05 → stationary

(b) KPSS Test

Null hypothesis (H₀): Series is stationary

If p-value < 0.05 → non-stationary

Using both tests together gives more reliable results.

**4. Transforming a Non-Stationary Series into Stationary:**
**4.1 Differencing**

Subtract current value from previous value

Removes trend

Example: $ yt′​=yt​−yt−1​ $
	​

- First-order differencing → removes linear trend

- Second-order differencing → removes complex trend

**4.2 Log / Power Transformation:**

Stabilizes variance

Useful when variance increases over time

**Examples:**

Log transformation

Square root transformation

4.3 Seasonal Differencing

Removes seasonal effects
$ yt′​=yt​−yt−s​ $

(where s = seasonal period)

**4.4 Detrending:**

- Remove trend using regression or smoothing

- Residual series becomes stationary

**4.5 Decomposition:**

Decompose series into Trend, Seasonal, Residual

Use residual component for modeling

**Example:**

A sales time series shows:

- Upward trend

- Increasing variance

- Seasonal spikes during festivals

Steps:

- Apply log transformation

- Use first-order differencing

- Apply seasonal differencing

- Test stationarity using ADF test

**Conclusion:** Stationarity is a fundamental requirement in time series analysis. A stationary series has constant mean, variance, and autocovariance over time. Non-stationary series can be tested using visual methods and statistical tests like ADF and KPSS, and transformed into stationary form using differencing, transformations, and decomposition techniques. Ensuring stationarity improves the accuracy and reliability of time-series forecasting models.

5. Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models in
terms of structure and application.

--> Time series forecasting is widely used in finance, economics, sales forecasting, weather prediction, and demand planning. Models such as AR, MA, ARIMA, SARIMA, and SARIMAX are classical statistical approaches used to model time-dependent data. These models differ in their structure, assumptions, and application areas.

1. **AutoRegressive (AR) Model** : The AutoRegressive (AR) model predicts the current value of a time series using its past values.

Structure:

$Yt​=c+ϕ1​Yt−1​+ϕ2​Yt−2​+⋯+ϕp​Yt−p​+ϵt​$

- Depends on p lagged observations

- Requires stationary data

**Application:**

- Stock prices

- Economic indicators

- Stationary financial time series

2. **Moving Average (MA) Model** : The Moving Average (MA) model predicts the current value using past forecast errors.

Structure:

$Yt​=c+θ1​ϵt−1​+θ2​ϵt−2​+⋯+θq​ϵt−q​$

- Captures short-term shock effects

- Suitable for stationary series

**Application:**

- Noise filtering

- Sensor readings

- Error-driven time series

3. **AutoRegressive Integrated Moving Average (ARIMA) Model** : ARIMA combines AR, MA, and differencing (I) to handle non-stationary time series.

Structure:

$
ARIMA(p,d,q)
$


- p → AR terms

- d → differencing order

- q → MA terms

**Application:**

- Sales forecasting

- Economic time series

- Demand prediction without seasonality

4. **Seasonal ARIMA (SARIMA) Model** :SARIMA extends ARIMA to model seasonal patterns in time series data.

Structure:

SARIMA(p,d,q)(P,D,Q,s)

Seasonal AR, MA, differencing

s = seasonal period (e.g., 12 for monthly data)

**Application:**

- Monthly sales forecasting

- Electricity demand

- Weather and climate data

5. **Seasonal ARIMA with Exogenous Variables (SARIMAX)** : SARIMAX further extends SARIMA by incorporating external (exogenous) variables that influence the time series.

Structure :

SARIMAX(p,d,q)(P,D,Q,s)+X

- Includes additional predictors like interest rates, promotions, inflation

**Application:**

- Sales affected by promotions

- Loan default prediction with economic indicators

- Traffic forecasting with weather data

**COMPARISON"**
| Model | Structure | Seasonality | External Variables | Application |
|------|-----------|-------------|--------------------|-------------|
| AR | Uses past values | No | No | Stationary financial data |
| MA | Uses past errors | No | No | Noise-driven data |
| ARIMA | AR + differencing + MA | No | No | Non-stationary series |
| SARIMA | ARIMA + seasonal terms | Yes | No | Seasonal time series |
| SARIMAX | SARIMA + exogenous variables | Yes | Yes | Time series with external influence |

**Real-World Example:**

1. In a retail business:

- Daily sales → ARIMA

- Monthly sales with festivals → SARIMA

- Monthly sales affected by promotions & discounts → SARIMAX


**Conclusion:** AR, MA, ARIMA, SARIMA, and SARIMAX models form a hierarchy of increasing complexity for time series forecasting. While AR and MA handle simple stationary data, ARIMA manages non-stationarity, SARIMA captures seasonality, and SARIMAX incorporates external variables. Choosing the right model depends on data characteristics such as stationarity, seasonality, and the presence of external influencing factors.

```python
# 6: Load a time series dataset (e.g., AirPassengers), plot the original series,
# and decompose it into trend, seasonality, and residual components

#CODE:

import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Load AirPassengers dataset
data = pd.read_csv(
    "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv",
    parse_dates=["Month"],
    index_col="Month"
)

# Plot original time series
plt.figure()
plt.plot(data, label="Passengers")
plt.title("AirPassengers Time Series")
plt.legend()
plt.show()

# Decompose the time series
decomposition = seasonal_decompose(data, model="multiplicative")
decomposition.plot()
plt.show()
```

```python
#  7: Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare) to
# detect anomalies. Visualize the anomalies on a 2D scatter plot.

#CODE:
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# Sample dataset
data = pd.DataFrame({
    "fare_amount": [5, 7, 8, 6, 9, 100, 6, 7, 8, 120],
    "trip_distance": [1, 2, 1.5, 1.2, 2.1, 30, 1.3, 1.4, 1.6, 35]
})

# Train Isolation Forest
model = IsolationForest(contamination=0.2, random_state=42)
data["anomaly"] = model.fit_predict(data)

# Plot
plt.figure()
plt.scatter(
    data["fare_amount"],
    data["trip_distance"],
    c=data["anomaly"]
)
plt.xlabel("Fare Amount")
plt.ylabel("Trip Distance")
plt.title("Isolation Forest Anomaly Detection")
plt.show()
```

```python
#8: Train a SARIMA model on the monthly airline passengers dataset.
#Forecast the next 12 months and visualize the results.

import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Reload AirPassengers dataset to ensure it's the correct univariate data
data = pd.read_csv(
    "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv",
    parse_dates=["Month"],
    index_col="Month"
)

# SARIMA model
model = SARIMAX(
    data,
    order=(1,1,1),
    seasonal_order=(1,1,1,12)
)
result = model.fit()

# Forecast next 12 months
forecast = result.get_forecast(steps=12)
forecast_values = forecast.predicted_mean

# Plot
plt.figure()
plt.plot(data, label="Actual")
plt.plot(forecast_values, label="Forecast")
plt.legend()
plt.title("SARIMA Forecast - Next 12 Months")
plt.show()
```

```python
#9. Apply Local Outlier Factor (LOF) on any numerical dataset to detect
#anomalies and visualize them using matplotlib.

#CODE:
from sklearn.neighbors import LocalOutlierFactor
import numpy as np
import matplotlib.pyplot as plt

# Sample numerical data
X = np.array([[1],[2],[2],[3],[2],[100],[2],[3]])

# LOF model
lof = LocalOutlierFactor(n_neighbors=2)
labels = lof.fit_predict(X)

# Plot
plt.figure()
plt.scatter(range(len(X)), X, c=labels)
plt.xlabel("Index")
plt.ylabel("Value")
plt.title("LOF Anomaly Detection")
plt.show()
```

10. You are working as a data scientist for a power grid monitoring company.
Your goal is to forecast energy demand and also detect abnormal spikes or drops in
real-time consumption data collected every 15 minutes. The dataset includes features
like timestamp, region, weather conditions, and energy usage.
Explain your real-time data science workflow:
- How would you detect anomalies in this streaming data (Isolation Forest / LOF /
DBSCAN)?

 - Which time series model would you use for short-term forecasting (ARIMA /
SARIMA / SARIMAX)?

-  How would you validate and monitor the performance over time?

- How would this solution help business decisions or operations?

--> # Real-Time Energy Demand Forecasting and Anomaly Detection Workflow

You are tasked with forecasting energy demand and detecting abnormal consumption in real-time for a power grid monitoring company. The dataset includes **timestamp, region, weather conditions, and energy usage**.

---

## 1️. Anomaly Detection in Streaming Data

**Goal:** Detect sudden spikes or drops in energy usage to prevent overloads or failures.

**Approach:**
- Use **Isolation Forest** for real-time anomaly detection
  - Works efficiently on **large, streaming datasets**
  - Does not require labeled anomalies
  - Isolates abnormal points quickly based on how easily they can be separated from normal data

**Alternative Methods:**
- **Local Outlier Factor (LOF):** Good for local density anomalies
- **DBSCAN:** Density-based, less suitable for continuous streaming

**Key Steps:**
1. Collect **15-min interval energy usage data**
2. Update the Isolation Forest model periodically (online or mini-batch)
3. Flag points with **anomaly score = -1** as abnormal

---

## 2️. Forecasting Model for Short-Term Demand

**Goal:** Forecast energy usage for the next 15–60 minutes to optimize grid operations.

**Approach:**
- Use **SARIMAX (Seasonal ARIMA with Exogenous variables)**
  - **Seasonality:** Captures daily and weekly patterns in energy usage
  - **Exogenous variables:** Incorporates weather conditions, region, temperature, humidity

**Model parameters example:**
- `order=(p,d,q)` → ARIMA parameters
- `seasonal_order=(P,D,Q,s)` → Seasonal patterns (e.g., `s=96` for 15-min intervals in 1 day)

**Alternative Models:**
- **ARIMA:** Only for non-seasonal, stationary series
- **SARIMA:** Seasonal but without exogenous variables

---

## 3️. Validation and Monitoring

**Goal:** Ensure model accuracy and detect anomalies reliably over time.

**Validation Metrics:**
- **MAE (Mean Absolute Error)** → Average error
- **RMSE (Root Mean Squared Error)** → Penalizes large deviations

**Monitoring:**
- Track **anomaly detection rate** over time
- Retrain model periodically as new data arrives
- Maintain **performance dashboards** for energy managers

**Additional Checks:**
- Visual inspection of predicted vs actual energy usage
- Compare detected anomalies with historical events (holidays, maintenance, weather events)

---

## 4️. Business Impact

Implementing this solution helps the company:

- **Prevent Power Outages:** Detect abnormal spikes in demand early
- **Optimize Energy Distribution:** Forecast demand to balance load across regions
- **Reduce Operational Costs:** Avoid emergency overproduction or blackout penalties
- **Enable Proactive Maintenance:** Identify unusual patterns in energy usage that indicate potential equipment issues


```python
#Python implementation

# Step 0: Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_absolute_error, mean_squared_error


# Step 1: Generate Sample Dataset
# 15-min interval timestamps for 2 days
timestamps = pd.date_range(start='2026-01-01', periods=192, freq='15T')

# Simulate energy usage with seasonality + random noise
np.random.seed(42)
base_usage = 50 + 10*np.sin(np.linspace(0, 4*np.pi, 192))  # daily seasonality
noise = np.random.normal(0, 2, 192)
energy_usage = base_usage + noise

# Simulate weather features
temperature = 20 + 5*np.sin(np.linspace(0, 4*np.pi, 192)) + np.random.normal(0, 1, 192)
humidity = 50 + 10*np.cos(np.linspace(0, 4*np.pi, 192)) + np.random.normal(0, 2, 192)

# Introduce anomalies (spikes/drops)
energy_usage[50] = 100  # spike
energy_usage[150] = 10  # drop

# Create DataFrame
data = pd.DataFrame({
    'timestamp': timestamps,
    'energy_usage': energy_usage,
    'temperature': temperature,
    'humidity': humidity
})
data.set_index('timestamp', inplace=True)


# Step 2: Anomaly Detection (Isolation Forest)
iso = IsolationForest(contamination=0.02, random_state=42)
data['anomaly'] = iso.fit_predict(data[['energy_usage']])
# 1 = normal, -1 = anomaly

# Visualize anomalies
plt.figure(figsize=(12,5))
plt.plot(data['energy_usage'], label='Energy Usage')
plt.scatter(data.index[data['anomaly']==-1], data['energy_usage'][data['anomaly']==-1],
            color='red', label='Anomaly', s=100)
plt.title('Energy Usage with Anomalies')
plt.xlabel('Timestamp')
plt.ylabel('Energy Usage')
plt.legend()
plt.show()


# Step 3: SARIMAX Forecasting
# Use last 168 points for training (exclude last 24 for testing)
train = data.iloc[:-24]
test = data.iloc[-24:]

# SARIMAX model with daily seasonality (96 intervals per day)
model = SARIMAX(
    train['energy_usage'],
    exog=train[['temperature','humidity']],
    order=(1,1,1),
    seasonal_order=(1,1,1,96)
)
result = model.fit(disp=False)

# Forecast next 24 steps using exogenous variables from test set
forecast = result.get_forecast(steps=24, exog=test[['temperature','humidity']])
predicted_values = forecast.predicted_mean
conf_int = forecast.conf_int()


# Step 4: Plot Forecast vs Actual
plt.figure(figsize=(12,5))
plt.plot(train.index, train['energy_usage'], label='Training Data')
plt.plot(test.index, test['energy_usage'], label='Actual Energy Usage')
plt.plot(test.index, predicted_values, label='SARIMAX Forecast', color='green')
plt.fill_between(test.index, conf_int.iloc[:,0], conf_int.iloc[:,1], color='green', alpha=0.2)
plt.scatter(data.index[data['anomaly']==-1], data['energy_usage'][data['anomaly']==-1],
            color='red', label='Anomalies', s=100)
plt.title('Energy Usage Forecast with Anomalies')
plt.xlabel('Timestamp')
plt.ylabel('Energy Usage')
plt.legend()
plt.show()


# Step 5: Evaluate Forecast Performance
mae = mean_absolute_error(test['energy_usage'], predicted_values)
# Calculate RMSE manually as 'squared' argument is not supported in older sklearn versions
mse = mean_squared_error(test['energy_usage'], predicted_values)
rmse = np.sqrt(mse)

print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")


# Step 6: Display Detected Anomalies
print("Detected Anomalies:")
print(data[data['anomaly']==-1])
```
