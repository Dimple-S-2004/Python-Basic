1. What is Simple Linear Regression (SLR)? Explain its purpose.

--> Simple Linear Regression (SLR) is a statistical and machine-learning technique used to model the relationship between one independent variable (X) and one dependent variable (Y) by fitting a straight line to the data.

Mathematical form: Y = Œ≤0 + Œ≤1X + Œµ

Œ≤0 : Intercept (value of Y when X = 0)

Œ≤1 : Slope (change in Y for 1 unit change in X)

Œµ  : Error term

*Purpose of Simple Linear Regression*:

* Understand relationship
To identify and quantify how one variable affects another (e.g., how salary changes with experience).

* Prediction
To predict the value of the dependent variable for a given value of the independent variable.

* Trend analysis
To observe patterns and trends in data using a straight-line approximation.

* Decision making
Helps in making data-driven decisions based on the strength and direction of the relationship.

Example:

Predicting house price (Y) based on house size (X)

Predicting marks (Y) based on hours studied (X)

2. What are the key assumptions of Simple Linear Regression?

--> Key Assumptions of Simple Linear Regression (SLR)

Simple Linear Regression works correctly only when the following assumptions are satisfied:

* Linearity
The relationship between the independent variable (X) and dependent variable (Y) is linear.

* Independence of errors
The error terms (residuals) are independent of each other (no autocorrelation).

* Homoscedasticity
The variance of the error terms is constant across all values of X.

* Normality of errors
The error terms are normally distributed with a mean of zero.

* No or minimal outliers
Extreme values should not strongly influence the regression line.

Simple Linear Regression assumes linearity, independence, constant variance, normal distribution of errors, and absence of influential outliers.

Explanation of Each Term:

* Independent Variable (X)

X is the predictor or input variable.

It is used to explain changes in the dependent variable Y.

* Intercept (Œ≤0)

Œ≤0 is the value of Y when X equals zero.

It indicates where the regression line intersects the Y-axis.

It provides the baseline value of the dependent variable.

 * Slope / Regression Coefficient (Œ≤1)

Œ≤1 represents the rate of change of Y with respect to X.

It shows how much Y increases or decreases when X increases by one unit.

A positive Œ≤1 indicates a positive relationship, while a negative Œ≤1 indicates an inverse relationship.

* Error Term (Œµ)

Œµ represents the random error in the model.

It accounts for the variation in Y that cannot be explained by X.

It includes effects of omitted variables, noise, and measurement errors.

5. What is the method of least squares in linear regression?

Y=Œ≤0‚Äã+Œ≤1‚ÄãX



For each data point, the model predicts a value Y^.The difference between the actual value ùëå and the predicted value Y^ is called the error or residual.

The method of least squares works by minimizing the sum of the squares of these residuals:

7. Name and briefly describe three common evaluation metrics for regression
models.

--> In regression, we predict continuous numerical values, and it is important to measure how accurate the predictions are. The most commonly used metrics are MAE, MSE, and R¬≤ score.

1. Mean Absolute Error (MAE)

Definition:
MAE calculates the average absolute difference between the actual values
ùëåi and predicted values Y^i.

$$
\text{MAE} = \frac{1}{n} \sum_{i=1}^{n} \big| Y_i - \hat{Y}_i \big|
$$

Interpretation:

* MAE is in the same units as the target variable.

* It treats all errors equally (no squaring), so it is less sensitive to outliers.

Use case:
Good when all errors are equally important, e.g., predicting house prices where every unit of error matters.

2. Mean Squared Error (MSE)

Definition:
MSE calculates the average of squared differences between actual and predicted values.

$$
\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (Y_i - \hat{Y}_i)^2
$$

Interpretation:

* Squaring the errors penalizes larger errors more heavily.

* Sensitive to outliers because large errors dominate the metric.

Use case:
Useful when large errors are particularly undesirable, e.g., predicting energy consumption or financial forecasts.

3. R-squared(R^2)Score

Definition:

ùëÖ^2 measures the proportion of variance in the dependent variable explained by the model.

$$
R^2 = 1 - \frac{\sum_{i=1}^{n} (Y_i - \hat{Y}_i)^2}{\sum_{i=1}^{n} (Y_i - \bar{Y})^2}
$$

Interpretation:

* Value ranges 0 to 1 (sometimes negative if the model is worse than predicting the mean).

* Higher R^2 indicates better model fit.

Use case:
Commonly used to compare different regression models and understand how well the model explains the data.









```python
# Simple code on Evaluation Metrics for Regression Models
# Import libraries
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Sample data
y_actual = [100, 150, 200, 250, 300]
y_pred = [110, 140, 210, 240, 310]

# Calculate metrics
mae = mean_absolute_error(y_actual, y_pred)
mse = mean_squared_error(y_actual, y_pred)
r2 = r2_score(y_actual, y_pred)

# Print results
print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared (R2 Score): {r2}")
```

```python
#9.Write Python code to fit a simple linear regression model using scikit-learn
#and print the slope and intercept.

# Import necessary libraries
from sklearn.linear_model import LinearRegression
import numpy as np

# Sample data
# X = independent variable (e.g., Area of house)
# y = dependent variable (e.g., House price)
X = np.array([1000, 1500, 2000, 2500, 3000]).reshape(-1, 1)  # Reshape to 2D array
y = np.array([150000, 200000, 250000, 300000, 350000])

# Create and fit the model
model = LinearRegression()
model.fit(X, y)

# Get slope and intercept
slope = model.coef_[0]
intercept = model.intercept_

# Print results
print(f"Slope (Œ≤1): {slope}")
print(f"Intercept (Œ≤0): {intercept}")
```

10. How do you interpret the coefficients in a simple linear regression model?

--> Interpreting Coefficients in Simple Linear Regression

In **Simple Linear Regression**, the model is expressed as:

$$
Y = \beta_0 + \beta_1 X
$$

Where:  
- $Y$ = Dependent variable (target)  
- $X$ = Independent variable (predictor)  
- $\beta_0$ = Intercept  
- $\beta_1$ = Slope (regression coefficient)  

The coefficients $\beta_0$ and $\beta_1$ represent the relationship between X and Y.


 1. Intercept ($\beta_0$)

- **Definition:**  
  The intercept is the **predicted value of Y when X = 0**.  

- **Interpretation:**  
  - Represents the **starting point of the regression line** on the Y-axis.  
  - Example: In a house price model, if $\beta_0 = 50,000$, it means a house with **0 sq. ft area** is estimated to cost ‚Çπ50,000.  
  - Sometimes, the intercept may **not have practical meaning** if X = 0 is unrealistic (e.g., adult age vs. salary).



 2. Slope ($\beta_1$)

- **Definition:**  
  The slope indicates **how much Y changes for a one-unit increase in X**.  

- **Interpretation:**  
  - Positive slope ($\beta_1 > 0$) ‚Üí Y increases as X increases  
  - Negative slope ($\beta_1 < 0$) ‚Üí Y decreases as X increases  
  - Example: In a house price model, if $\beta_1 = 3000$, it means **for every additional 1 sq. ft of area, the house price increases by ‚Çπ3,000**.  

- **Importance:**  
  - Shows the **strength and direction** of the relationship between X and Y.  
  - Helps in **prediction** and **decision-making**.


 3. General Interpretation Steps

1. Check **sign of $\beta_1$** ‚Üí positive or negative relationship  
2. Look at **magnitude of $\beta_1$** ‚Üí how much Y changes per unit of X  
3. Consider **units of X and Y** ‚Üí interpret in real-world terms  
4. Consider **context** ‚Üí if X = 0 is unrealistic, the intercept may be theoretical  



4. Example

Suppose the regression equation is:

$$
\text{House Price} = 50,000 + 3,000 \times \text{Area}
$$

- **Intercept ($\beta_0 = 50,000$)** ‚Üí Base price of a house when area is 0  
- **Slope ($\beta_1 = 3,000$)** ‚Üí Price increases by ‚Çπ3,000 per sq. ft  

Interpretation in real-world terms:  
> ‚ÄúA house with zero area is valued at ‚Çπ50,000 (baseline). For every additional 1 sq. ft of area, the house price increases by ‚Çπ3,000.‚Äù



 5. Key Points

- **Intercept** ‚Üí baseline/predicted value when X = 0  
- **Slope** ‚Üí effect of X on Y; indicates direction and strength  
- Always relate to **real-world meaning**  
- Use slope for **prediction** and decision-making  



**Conclusion:**  
The coefficients in a simple linear regression model provide quantitative insights into the relationship between variables, help in prediction, and are essential for understanding how changes in the predictor affect the outcome.


```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data (Independent variable X and Dependent variable Y)
X = np.array([1000, 1500, 2000, 2500, 3000]).reshape(-1, 1)
Y = np.array([150000, 200000, 250000, 300000, 350000])

# Create and fit Linear Regression model
model = LinearRegression()
model.fit(X, Y)

# Get coefficients
slope = model.coef_[0]
intercept = model.intercept_

# Print slope and intercept
print(f"Slope (Œ≤1): {slope}")
print(f"Intercept (Œ≤0): {intercept}")

# Predict values for plotting
Y_pred = model.predict(X)

# Plot data points
plt.scatter(X, Y, color='blue', label='Actual Data')
plt.plot(X, Y_pred, color='red', label='Regression Line')

# Annotate slope and intercept
plt.text(1200, 160000, f"Slope (Œ≤1) = {slope:.2f}", color='green')
plt.text(1200, 155000, f"Intercept (Œ≤0) = {intercept:.2f}", color='purple')

# Labels and title
plt.xlabel('Area (sq. ft)')
plt.ylabel('House Price (‚Çπ)')
plt.title('Simple Linear Regression: House Price vs Area')
plt.legend()
plt.show()
```
