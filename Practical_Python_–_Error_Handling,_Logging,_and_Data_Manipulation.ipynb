1. What is the difference between Multithreading and Multiprocessing?
| **Feature**                       | **Multithreading**                                                         | **Multiprocessing**                                                    |
| --------------------------------- | -------------------------------------------------------------------------- | ---------------------------------------------------------------------- |
| **Definition**                    | Runs multiple **threads** (lightweight processes) within the same process. | Runs multiple **independent processes** simultaneously.                |
| **Memory**                        | Threads **share the same memory space**.                                   | Each process has its **own separate memory space**.                    |
| **Speed**                         | Faster for **I/O-bound tasks** (e.g., file reading, network calls).        | Faster for **CPU-bound tasks** (e.g., heavy computations).             |
| **Communication**                 | **Easier**, since threads share memory directly.                           | **Slower**, requires **Inter-Process Communication (IPC)** mechanisms. |
| **GIL (Global Interpreter Lock)** | **Affected** — only one thread executes Python code at a time.             | **Not affected** — each process runs in its own Python interpreter.    |
| **Example Use**                   |  Web scraping, I/O operations, background tasks.                           |  Image processing, data computation, machine learning workloads.        |

2. What are the challenges associated with memory management in Python?

Memory management in Python is mostly automatic, but certain issues can still arise due to how Python handles references and garbage collection.
Main Challenges

Garbage Collection Overhead: Automatic garbage collector may cause performance drops.

Reference Cycles: Objects referencing each other can delay memory release.

Memory Fragmentation: Frequent allocation/deallocation can fragment memory.

Large Object Handling: Large datasets consume a lot of memory without optimization.

Global Interpreter Lock(GIL): Limits true parallelism for CPU tasks.
Inefficient Data Structures: Using lists or dictionaries for huge numeric data can waste memory compared to specialized structures (like NumPy arrays).


```python
#3. Write a Python program that logs an error message to a log file when a division by zero exception occurs.
#The program demonstrates exception handling and logging using Python’s logging module.
#If a division by zero happens, the program logs the error in a file named error.log.


import logging   # Configure logging
logging.basicConfig(filename='error.log', level=logging.ERROR)

try:
    a = 10
    b = 0
    result = a / b
    print("Result:", result)
except ZeroDivisionError as e:
    logging.error("Division by zero error: %s", e)
    print("Error: Cannot divide by zero.")

```

```python
#4.Write a Python program that reads from one file and writes its content to another file.
#	This program demonstrates file handling — reading data from one file and writing it into another.

# Create a new file
with open("source.txt", "w") as f:
    f.write("Hello! This is the content of the source   file.\n")
    f.write("This will be copied to another file.")
# Create an empty file
open("source.txt", "w").close()
# Open the file in append mode
with open("source.txt", "a") as f:
    f.write("\nNew line added later.")
with open("source.txt", "r") as src:
    content = src.read()

with open("destination.txt", "w") as dest:
    dest.write(content)

print("File content copied successfully!")

```

```python
#5.Write a program that handles both IndexError and KeyError using a try-except block.
#	Python’s try-except block can handle multiple exception types.
#Here, we handle IndexError (list index out of range) and KeyError (missing dictionary key).
try:
    my_list = [1, 2, 3]
    my_dict = {'a': 10, 'b': 20}

    print(my_list[5])     # May cause IndexError
    print(my_dict['z'])   # May cause KeyError

except IndexError:
    print("Index out of range error.")
except KeyError:
    print("Key not found in dictionary.")
```

6.What are the differences between NumPy arrays and Python lists?

| **Feature**                 | **Python List**                                                  | **NumPy Array**                                           |
| --------------------------- | ---------------------------------------------------------------- | --------------------------------------------------------- |
| **Data Type**               | Can store elements of **different types**.                       | Stores elements of the **same data type** only.           |
| **Memory Efficiency**       | **Consumes more memory**.                                        | **More memory-efficient** due to homogeneous storage.     |
| **Speed**                   | **Slower** for numerical operations.                             | **Faster** (implemented in optimized C code).             |
| **Mathematical Operations** | Requires **explicit loops** for element-wise operations.         | Supports **vectorized operations** (no explicit loops).   |
| **Dimensionality**          | Primarily **1D**, needs nested lists for multi-dimensional data. | Supports **multi-dimensional arrays** (1D, 2D, 3D, etc.). |
| **Functionality**           | **Limited** built-in math operations.                            | Offers **many mathematical and statistical functions**.   |

7.Explain the difference between apply() and map() in Pandas.
| **Feature**              | **`map()`**                                        | **`apply()`**                                                |
| ------------------------ | -------------------------------------------------- | ------------------------------------------------------------ |
| **Used On**              | Works on a **Series** (single column).             | Works on a **DataFrame or Series** (entire dataset).         |
| **Function Application** | Applies a function **element-wise** to each value. | Applies a function **along an axis** (rows or columns).      |
| **Return Type**          | Always returns a **Series**.                       | Returns a **Series or DataFrame** depending on the function. |
| **Scope**                | Works only on **1D data (Series)**.                | Works on **1D or 2D data (Series or DataFrame)**.            |


```python
#8. Create a histogram using Seaborn to visualize a distribution.
#	A histogram shows the frequency distribution of a dataset — how values are spread over intervals.

import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = [10, 20, 22, 24, 25, 25, 26, 28, 29, 30, 35, 40, 42, 43, 45]

# Set smaller figure size (width=5, height=3)
plt.figure(figsize=(5, 3))

# Create histogram
sns.histplot(data, bins=8, kde=True, color='skyblue')

plt.title("Distribution Histogram", fontsize=10)
plt.xlabel("Values", fontsize=9)
plt.ylabel("Frequency", fontsize=9)
plt.tight_layout()
plt.show()
```

```python
#10.Calculate the correlation matrix using Seaborn and visualize it with a heatmap.
	#A correlation matrix shows the relationship between numeric columns, and a heatmap visualizes it using colors.

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = {
    'Math': [78, 85, 92, 88, 76, 95, 89],
    'Science': [82, 79, 91, 85, 80, 94, 88],
    'English': [75, 80, 85, 78, 82, 90, 84]
}

# Create DataFrame
df = pd.DataFrame(data)

# Calculate correlation matrix
corr = df.corr()

# Display correlation matrix using heatmap
plt.figure(figsize=(5, 3))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
