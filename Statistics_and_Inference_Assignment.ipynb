1.What are Type I and Type II errors in hypothesis testing, and how do they
impact decision-making?

--> In hypothesis testing, we aim to determine if there is enough evidence to reject a null hypothesis ($H_0$) in favor of an alternative hypothesis ($H_a$). However, there is always a risk of making an incorrect decision. These incorrect decisions are known as Type I and Type II errors.

*   **Type I Error (False Positive):** This occurs when we reject the null hypothesis ($H_0$) when it is actually true. It's like saying there is an effect or difference when there isn't one. The probability of making a Type I error is denoted by $\alpha$ (alpha) and is often referred to as the significance level.

*   **Type II Error (False Negative):** This occurs when we fail to reject the null hypothesis ($H_0$) when it is actually false. It's like saying there is no effect or difference when there actually is one. The probability of making a Type II error is denoted by $\beta$ (beta).

**Impact on Decision-Making:**

The impact of these errors depends on the context of the hypothesis test.

*   **Type I Error:** Can lead to unnecessary changes, wasted resources, or incorrect conclusions. For example, in a medical trial, a Type I error could mean concluding a drug is effective when it's not, leading to its approval and use with no real benefit.

*   **Type II Error:** Can lead to missed opportunities or a failure to detect a real effect or difference. For example, in a medical trial, a Type II error could mean failing to detect that a drug is effective when it is, preventing it from being used to help patients.

**Comparison Between Type I and Type II Errors**

| Aspect          | Type I Error                       | Type II Error                           |
|-----------------|------------------------------------|-----------------------------------------|
| Definition      | Rejecting a true null hypothesis   | Failing to reject a false null hypothesis |
| Symbol          | $\alpha$ (alpha)                   | $\beta$ (beta)                          |
| Nature          | False Positive                     | False Negative                          |
| Example         | Concluding a new drug works when it doesn't | Concluding a new drug doesn’t work when it actually does |
| Consequence     | Leads to false alarms              | Leads to missed detections             |
| Control Method  | Set significance level ($\alpha$)  | Increase sample size or test power     |

**Conclusion**

Understanding Type I and Type II errors is crucial in hypothesis testing. The choice of which error to minimize depends on the specific situation and the relative costs of each type of error. There is often a trade-off between reducing one type of error and increasing the other.

2.What is the P-value in hypothesis testing, and how should it be interpreted
in the context of the null hypothesis?

--> P-Value in Hypothesis Testing:

In hypothesis testing, the P-value plays a crucial role in deciding whether to reject or fail to reject the null hypothesis (H₀). It provides a measure of the strength of evidence against the null hypothesis based on the sample data.

1. Definition

The P-value (probability value) is the probability of obtaining test results at least as extreme as the results actually observed, assuming that the null hypothesis is true.

In simple terms, it tells us how likely our sample results are if the null hypothesis were correct.

2. Interpretation

A small P-value indicates strong evidence against the null hypothesis.
A large P-value indicates weak evidence against the null hypothesis.
The P-value is compared with a pre-decided significance level ( α ), usually 0.05 or 5%.

3. Decision Rule

Condition	Decision	Interpretation
P-value  ≤α 	Reject H₀	There is enough evidence to support the alternative hypothesis (H₁).
P-value  >α 	Fail to Reject H₀	There is not enough evidence to reject the null hypothesis.

4. Example : Suppose a researcher tests whether a new medicine is more effective than the current one.

Null Hypothesis (H₀): Both medicines are equally effective.
Alternative Hypothesis (H₁): The new medicine is more effective.
If the test gives a P-value = 0.02 and  α  = 0.05,

Since 0.02 < 0.05, we reject H₀, meaning there is significant evidence that the new medicine is more effective.

3:Explain the difference between a Z-test and a T-test, including when to use
each.

-->

| **Feature** | **Z-test** | **T-test** |
|--------------|-------------|-------------|
| **Definition** | A statistical test used to determine whether there is a significant difference between sample and population means when the population variance is known. | A statistical test used to determine whether there is a significant difference between sample and population means when the population variance is **unknown**. |
| **Population Standard Deviation (σ)** | Known | Unknown (estimated from sample) |
| **Sample Size** | Large sample (n ≥ 30) | Small sample (n < 30) |
| **Test Statistic Formula** | Z = ( X̄ - μ ) / ( σ / √n ) | t = ( X̄ - μ ) / ( s / √n ) |
| **Distribution Used** | Standard Normal Distribution (Z-distribution) | Student’s T-Distribution |
| **Shape of Distribution** | Symmetrical and narrower | Symmetrical but wider (heavier tails) |
| **When to Use** | When population variance is known and sample size is large | When population variance is unknown and sample size is small |

 **In short:**  
- Use **Z-test** when σ (population standard deviation) is **known** and sample is **large**.  
- Use **T-test** when σ is **unknown** and sample is **small**.

4:What is a confidence interval, and how does the margin of error influence its width and interpretation?

--> **Confidence Interval (CI):**  
A confidence interval gives a range of values that is likely to contain the true population parameter (like the mean or proportion) with a certain level of confidence (e.g., 95% or 99%).

**Formula (for mean):**  
CI = X̄ ± Margin of Error  

where  
Margin of Error = Z × (σ / √n)   or   t × (s / √n)


**Interpretation Example:**  
If the 95% confidence interval for the mean salary is ₹45,000 to ₹55,000, it means we are **95% confident** that the true average salary lies between ₹45,000 and ₹55,000.


**Margin of Error Influence:**

| **Factor** | **Effect on CI Width** | **Explanation** |
|-------------|------------------------|----------------|
| **Higher Margin of Error** | **Wider CI** | Indicates more uncertainty in the estimate. |
| **Lower Margin of Error** | **Narrower CI** | Indicates more precision in the estimate. |
| **Larger Sample Size (n)** | **Smaller Margin of Error** | Makes CI narrower and more precise. |
| **Higher Confidence Level (e.g., 99%)** | **Larger Margin of Error** | CI becomes wider because more confidence requires a broader range. |


 **In summary:**  
A **confidence interval** gives an estimated range for the true population value.  
The **margin of error** controls how **wide** or **narrow** that range is — higher precision (smaller margin) gives a **narrower interval**, while greater confidence requires a **wider interval**.

5: Describe the purpose and assumptions of an ANOVA test. How does it
extend hypothesis testing to more than two groups?

--->
**Purpose:**  
ANOVA (Analysis of Variance) is a statistical method used to determine whether there are any statistically significant differences between the means of **three or more independent groups**.

It helps extend hypothesis testing beyond two groups (which a t-test handles).  
While a t-test compares the means of **two groups**, ANOVA compares **multiple group means simultaneously**, reducing the risk of **Type I error** (false positives).



#### **Key Idea:**
ANOVA tests whether the variability **between group means** is greater than the variability **within groups**.

- **Null Hypothesis (H₀):** All group means are equal.  
- **Alternative Hypothesis (H₁):** At least one group mean is different.



### **Assumptions of ANOVA:**

| **Assumption** | **Description** |
|----------------|-----------------|
| **1. Independence** | The samples are independent of each other. |
| **2. Normality** | Each group is approximately normally distributed. |
| **3. Homogeneity of Variance** | All groups have roughly equal variances. |



### **How ANOVA Extends Hypothesis Testing:**
- **T-test:** compares **2 means** only.  
- **ANOVA:** can compare **3 or more means** at once.  
If ANOVA shows a significant difference, post-hoc tests (like **Tukey’s HSD**) can identify which specific groups differ.



 **Example:**  
Suppose we have three teaching methods (A, B, C) and we want to test if student's average scores differ among them.  
ANOVA helps check whether any method performs significantly differently from the others.



```python
#6. Write a Python program to perform a one-sample Z-test and interpret the
#result for a given dataset.
import numpy as np
from scipy import stats

# Sample data
data = [52, 47, 55, 49, 50, 51, 53, 48, 54, 46]
population_mean = 50
population_std = 5  # assumed known

# Calculate sample mean and size
sample_mean = np.mean(data)
n = len(data)

# Z-test statistic
z = (sample_mean - population_mean) / (population_std / np.sqrt(n))

# p-value (two-tailed)
p_value = 2 * (1 - stats.norm.cdf(abs(z)))

print("Z statistic:", z)
print("p-value:", p_value)
```

```python
#8. Generate multiple samples from a non-normal distribution and implement
#the Central Limit Theorem using Python.
import numpy as np
import matplotlib.pyplot as plt

# Generate data from a non-normal distribution (Exponential)
population = np.random.exponential(scale=2, size=100000)

sample_means = []

# Take 1000 samples of size 50 each
for _ in range(1000):
    sample = np.random.choice(population, size=50)
    sample_means.append(np.mean(sample))

# Plot original data and sample means
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.hist(population, bins=30, color='orange', edgecolor='black')
plt.title("Original Non-Normal Distribution (Exponential)")

plt.subplot(1, 2, 2)
plt.hist(sample_means, bins=30, color='skyblue', edgecolor='black')
plt.title("Sampling Distribution of Means (CLT Effect)")

plt.show()
```

```python
#9: Write a Python function to calculate and visualize the confidence interval
#for a sample mean.
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def plot_confidence_interval(data, confidence=0.95):
    # Sample statistics
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)
    n = len(data)

    # Standard error
    se = sample_std / np.sqrt(n)

    # Z critical value for given confidence level
    z_value = stats.norm.ppf((1 + confidence) / 2)

    # Confidence interval
    margin_of_error = z_value * se
    lower = sample_mean - margin_of_error
    upper = sample_mean + margin_of_error

    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"{int(confidence*100)}% Confidence Interval: ({lower:.2f}, {upper:.2f})")

    # Visualization
    plt.figure(figsize=(8, 4))
    plt.axvline(sample_mean, color='blue', label='Sample Mean')
    plt.axvline(lower, color='red', linestyle='--', label='Lower Bound')
    plt.axvline(upper, color='red', linestyle='--', label='Upper Bound')
    plt.title(f"{int(confidence*100)}% Confidence Interval for Sample Mean")
    plt.legend()
    plt.show()

# Example data
data = [45, 47, 49, 50, 52, 54, 55, 56, 58, 60]
plot_confidence_interval(data)
```

```python
#10.Perform a Chi-square goodness-of-fit test using Python to compare
#observed and expected distributions, and explain the outcome.
import numpy as np
from scipy.stats import chisquare

# Observed and expected frequencies
observed = np.array([25, 30, 20, 25])
expected = np.array([25, 25, 25, 25])  # uniform expectation

# Perform Chi-square test
chi_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print("Chi-square Statistic:", chi_stat)
print("p-value:", p_value)
```

### **Interpretation Example:**

If the p-value < 0.05:  
→ Reject H₀ → The observed distribution **differs significantly** from the expected one.

If the p-value ≥ 0.05:  
→ Fail to reject H₀ → The observed data **fits** the expected distribution well.

**Example Output:**
- Chi-Square Statistic = 2.5  
- p-value = 0.286 → Fail to reject H₀  

**Conclusion:**  
There is **no significant difference** between the observed and expected frequencies.  
Hence, the observed data **fits the expected distribution** according to the Chi-square goodness-of-fit test.
