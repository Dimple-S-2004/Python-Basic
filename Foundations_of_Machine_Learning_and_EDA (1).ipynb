1. What is the difference between AI, ML, DL, and Data Science? Provide a
brief explanation of each.

->Artificial Intelligence (AI), Machine Learning (ML), Deep Learning (DL), and Data Science are related fields, but each has a different scope and purpose. They work together, yet they are not the same. The explanation below clearly compares all four.

1. Artificial Intelligence (AI)
AI is the broadest field. It focuses on creating machines that can perform tasks that usually require human intelligence, such as reasoning, decision-making, and understanding language.

Scope:
Covers everything from simple rule-based systems to advanced autonomous systems.

Techniques Used:
- Rule-based logic
- Search and optimization algorithms
- NLP (Natural Language Processing)
- Expert systems
- Robotics
- Machine Learning and Deep Learning

Applications:
Virtual assistants, self-driving cars, chatbots, medical diagnosis systems, fraud detection, recommendation systems.

2. Machine Learning (ML)
ML is a subset of AI. It focuses on enabling machines to learn patterns from data and improve without explicit programming.

Scope:
Deals only with learning from historical data.

Techniques Used:
- Supervised learning (classification, regression)
- Unsupervised learning (clustering, dimensionality reduction)
- Reinforcement learning
- Algorithms like Decision Trees, Random Forests, SVM, KNN

Applications:
Spam detection, credit scoring, recommendation engines, customer churn prediction, stock trend analysis.

3. Deep Learning (DL)
DL is a subset of ML that uses neural networks with many layers (deep networks). It can automatically extract features from raw data, making it powerful for image, audio, and text tasks.

Scope:
Focuses on deep neural networks and requires large datasets and high computational resources.

Techniques Used:
- Convolutional Neural Networks (CNN)
- Recurrent Neural Networks (RNN)
- LSTM and GRU networks
- Transformers
- Autoencoders

Applications:
Image recognition, face detection, speech-to-text, translation, driver-assistance systems, chatbots like ChatGPT, medical image analysis.

4. Data Science
Data Science is a broader data-focused field. It involves collecting, cleaning, analyzing, and interpreting data to generate insights. It includes statistics, programming, visualization, and ML.

Scope:
Covers the entire data lifecycle from collection to insight generation.

Techniques Used:
- Statistical analysis
- Data cleaning and preprocessing
- Machine Learning
- Data visualization tools
- SQL and database querying
- Experimentation and hypothesis testing

Applications:
Business dashboards, sales forecasting, customer behavior analysis, A/B testing, marketing insights, fraud detection, operational optimization.

Comparison Table:

Aspect        AI                     ML                     DL                     Data Science
Scope         Broadest field         Subset of AI           Subset of ML           Broad data field
Goal          Build intelligent      Learn from data        Learn complex patterns Extract insights
              systems                                       via neural networks    from data
Data Need     Medium to High         Medium                 Very High              Varies
Techniques    Rules + ML + DL        ML algorithms          Neural Networks        Statistics + ML + Viz
Output        Intelligent behavior   Predictions            High accuracy models   Insights, reports

Conclusion:
AI is the broadest field that aims to create intelligent machines. ML is a way to achieve AI by learning from data. DL is an advanced part of ML that uses deep neural networks. Data Science focuses on analyzing data and extracting insights, using ML as one of its tools. Together, all four fields form the backbone of modern intelligent and data-driven systems.

2: Explain overfitting and underfitting in ML. How can you detect and prevent
them?

--> In Machine Learning, the goal is to build a model that performs well on both training data and unseen test data.
Two major problems that affect model performance are overfitting and underfitting, which are related to the bias–variance tradeoff.

*  Overfitting
Definition
Overfitting occurs when the model learns the training data too well, including noise and random fluctuations.
As a result, it performs very well on training data but poorly on test data.
Causes


Model is too complex

Too many features

Too many parameters


Deep trees or large neural networks


Example
A decision tree that splits until it perfectly classifies the training set but fails on new data.

*  Underfitting
Definition
Underfitting happens when the model is too simple and fails to capture the underlying patterns in the data.
It performs poorly on both training and test data.
Causes


Model has high bias


Not enough features


Too simple algorithm (e.g., using linear regression for nonlinear data)


Example
Fitting a straight line to a dataset that clearly has a curved pattern.

*  Bias–Variance Tradeoff


High Bias → Underfitting
Model is too simple and does not learn enough patterns.


High Variance → Overfitting
Model is too complex and memorizes the training data.


The goal is to find a balance between bias and variance for optimal performance.

*  How to Detect Overfitting / Underfitting
Detecting Overfitting


High training accuracy


Low test accuracy


Training loss decreases while validation loss increases


Detecting Underfitting


Training accuracy is low


Test accuracy is also low


Model fails to fit even the training data



*  How to Prevent Overfitting


Cross-validation (e.g., k-fold)


Regularization (L1, L2, Dropout)


Early stopping during training


Reduce model complexity


Data augmentation (for images)


Increase training data


*  How to Prevent Underfitting


Increase model complexity


Train for more epochs


Reduce regularization


Add more relevant features


Use more powerful algorithms



 *Conclusion*
Overfitting occurs when a model is too complex and memorizes noise, while underfitting happens when a model is too simple and fails to learn patterns.
Managing these issues through regularization, cross-validation, proper model complexity, and feature engineering ensures better generalization to new data.

3.How would you handle missing values in a dataset? Explain at least three
methods with examples.

--> Handling Missing Values in a Dataset
Missing values are a common issue in real-world datasets. They must be handled carefully because they can affect model accuracy and bias results. There are several methods to deal with missing data depending on the size of the dataset and the importance of the missing feature.

Method 1: Deletion (Removing Rows or Columns)

a) Row Deletion

If only a small number of rows have missing values, they can be removed.

Example:
A dataset of 5,000 rows has 10 rows where age is missing.
Removing these 10 rows will not affect overall accuracy.

b) Column Deletion

If a feature is missing in more than 60–70% of the rows, that feature may be dropped.

Example:
A “Twitter Username” column has 80% missing values.
Since it adds no major value, we drop the entire column.

When to use:

Missing values are very few

The column is not important



---

Method 2: Mean/Median Imputation

This method replaces missing values with a central value.

Mean Imputation (for numerical data)

Replace the missing value with the average value of the column.

Example:
Age column: [20, 25, 30, missing, 35]
Mean = (20+25+30+35)/4 = 27.5
Impute missing value with 27.5.

Median Imputation (for skewed data)

If data is skewed (like income), median is better.

Example:
Income: [10,000, 12,000, 15,000, missing, 5,00,000]
Median replaces the missing value safely without being influenced by the extreme outlier.

When to use:

Data is numeric

Missing values are moderate



---

Method 3: Predictive Modeling (Advanced Imputation)

Use ML models to predict missing values based on other features.

How it works:

Treat the missing feature as the target

Build a model using other columns

Predict the missing values


Example:
Missing “Salary” can be predicted using:
Age, Experience, Education, Job Role, etc.

Common models:

Regression (for numerical values)

Classification (for categorical values)

KNN Imputer (predicts using nearest neighbors)


When to use:

Important feature

Large dataset

Missing pattern is not random

4.What is an imbalanced dataset? Describe two techniques to handle it

-->An imbalanced dataset is a dataset where the distribution of target classes is not equal.
Example:

Class 0 = 95%

Class 1 = 5%

Machine learning models become biased toward the majority class, leading to poor performance on the minority class.

Techniques to Handle Imbalanced Data (Two Techniques)
1. SMOTE (Synthetic Minority Oversampling Technique)
Theory:

SMOTE creates synthetic samples of the minority class instead of simple duplication.
It works by:

Selecting a minority sample

Finding its K nearest minority neighbors

Generating new synthetic points between them

This increases minority class size and improves model balance.

2. Random Undersampling / Oversampling
Theory:

Random Oversampling: Duplicate existing minority samples to increase their count.
Random Undersampling: Remove random samples from the majority class to match minority count.
These methods are simple but can cause overfitting (oversampling) or loss of information (undersampling).

```python

# 1. Create a simple imbalanced datase
from sklearn.datasets import make_classification
from collections import Counter

X, y = make_classification(
    n_samples=500,
    n_features=2,
    n_redundant=0,
    n_clusters_per_class=1,
    weights=[0.90, 0.10],   # 90% class 0, 10% class 1 → imbalanced
    random_state=42
)

print("Original Class Distribution:", Counter(y))


# 2. Random Oversampling
from imblearn.over_sampling import RandomOverSampler

ros = RandomOverSampler(random_state=42)
X_over, y_over = ros.fit_resample(X, y)
print("After Random Oversampling:", Counter(y_over))


# 3. Random Undersampling
from imblearn.under_sampling import RandomUnderSampler

rus = RandomUnderSampler(random_state=42)
X_under, y_under = rus.fit_resample(X, y)
print("After Random Undersampling:", Counter(y_under))



# 4. SMOTE (Synthetic Minority Oversampling)
from imblearn.over_sampling import SMOTE

sm = SMOTE(random_state=42)
X_sm, y_sm = sm.fit_resample(X, y)
print("After SMOTE:", Counter(y_sm))

```

 5: Why is feature scaling important in ML? Compare Min-Max scaling and
Standardization.

-->Feature scaling is important because many ML algorithms are sensitive to the scale of numerical features. If features have very different ranges (e.g., Age: 0–100, Salary: 0–100,000), the model may give more importance to larger-scale features, even if they are not truly more important.

Algorithms affected by scaling:

* Distance-based models: KNN, K-Means, SVM → distances become distorted

* Gradient Descent models: Linear regression, logistic regression, neural networks → faster and stable convergence

* Regularization methods: Lasso and Ridge → penalty depends on magnitude

So scaling makes training faster, stable, and more accurate.

Min–Max Scaling (Normalization)

Formula:X_scaled = (X - X_min) / (X_max - X_min)

* Output range: 0 to 1
* Good for: Neural Networks, KNN, K-Means
* Sensitive to outliers.

Standardization (Z-score Scaling)

Formula:X_scaled = (X - mean) / standard_deviation

* Output range: mean = 0, std = 1
* Good for: SVM, Logistic Regression, Linear Regression
* Not affected much by outliers.

---

```python
# Q5 Feature Scaling: Min-Max vs Standardization

import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Creating a simple dataset
data = pd.DataFrame({
    'Age': [20, 25, 30, 35, 40, 45],
    'Salary': [20000, 30000, 40000, 60000, 80000, 100000]
})
print("Original Data:")
display(data)

# Min-Max Scaling
minmax = MinMaxScaler()
data_minmax = pd.DataFrame(minmax.fit_transform(data), columns=data.columns)
print("\nMin-Max Scaled Data (0–1 range):")
display(data_minmax)

# Standardization (Z-score)
std = StandardScaler()
data_std = pd.DataFrame(std.fit_transform(data), columns=data.columns)
print("\nStandardized Data (mean 0, std 1):")
display(data_std)
```

6. Compare Label Encoding and One-Hot Encoding. When would you prefer
one over the other?

-->1. Label Encoding

Label Encoding converts each category into a unique integer value.

Example:
Small → 0, Medium → 1, Large → 2

When to Use Label Encoding

 When the categorical variable has a natural order (ordinal data)
Examples:

Size (Small < Medium < Large)

Education (High School < Bachelor < Master < PhD)

Advantages

Simple and fast

Does not increase dataset dimensionality

Disadvantages

Creates an artificial numeric order for nominal categories

May mislead algorithms that rely on distance (KNN, SVM)

2. One-Hot Encoding

One-Hot Encoding converts categories into binary columns (0/1).

Example:
Red → [1,0,0], Blue → [0,1,0], Green → [0,0,1]

When to Use One-Hot Encoding

 When the categorical variable is nominal (no order)
Examples:

Color

City

Gender

Airline

Country

Advantages

No mistaken order between categories

Works well with distance-based models (KNN, SVM)

Disadvantages

Increases number of columns (high dimensionality)

Can cause the “curse of dimensionality” for large category counts

*Final Answer Statement*:

*Use Label Encoding when categories have meaningful order.
*Use One-Hot Encoding when categories are unordered and you want to avoid creating artificial relationships.

```python
#7:  Google Play Store Dataset
#a). Analyze the relationship between app categories and ratings. Which categories have the
#highest/lowest average ratings, and what could be the possible reasons?

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("googleplaystore.csv")


# Cleaning: Remove missing ratings
df = df.dropna(subset=["Rating"])

# a) Average Rating by Category
cat_ratings = df.groupby("Category")["Rating"].mean().sort_values()
print("Average Rating of Each Category:")
print(cat_ratings)

# Plot
plt.figure(figsize=(12,6))
sns.barplot(x=cat_ratings.index, y=cat_ratings.values)
plt.xticks(rotation=90)
plt.title("Average Rating by App Category")
plt.ylabel("Average Rating")
plt.xlabel("App Category")
plt.tight_layout()
plt.show()

```

* High-rated categories :
Education, Books, Health, and Productivity apps score higher because they solve real problems, offer useful content, and have fewer bugs—leading to better user satisfaction.

* Low-rated categories :
Games, Tools, Family, and Entertainment apps rate lower due to many low-quality or ad-heavy apps, frequent crashes, and high user expectations not being met.

```python
 #8: Titanic Dataset
#a) Compare the survival rates based on passenger class (Pclass). Which class had the highest
#survival rate, and why do you think that happened?
#b) Analyze how age (Age) affected survival. Group passengers into children (Age < 18) and
#adults (Age ≥ 18). Did children have a better chance of survival?

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("/content/titanic.csv")


# a) Survival Rate by Pclass

pclass_survival = df.groupby("Pclass")["Survived"].mean()
print("Survival Rate by Passenger Class:")
print(pclass_survival)

# Plot
plt.figure(figsize=(6,4))
sns.barplot(x=pclass_survival.index, y=pclass_survival.values)
plt.title("Survival Rate by Passenger Class")
plt.xlabel("Passenger Class (Pclass)")
plt.ylabel("Survival Rate")
plt.show()



# b) Survival based on Age Group (Child vs Adult)

# Remove missing ages
df = df.dropna(subset=["Age"])

# Create age group
df["AgeGroup"] = df["Age"].apply(lambda x: "Child" if x < 18 else "Adult")

age_survival = df.groupby("AgeGroup")["Survived"].mean()
print("\nSurvival Rate by Age Group:")
print(age_survival)

# Plot
plt.figure(figsize=(6,4))
sns.barplot(x=age_survival.index, y=age_survival.values)
plt.title("Survival Rate: Children vs Adults")
plt.xlabel("Age Group")
plt.ylabel("Survival Rate")
plt.show()

```

```python
#9: Flight Price Prediction Dataset
#a) How do flight prices vary with the days left until departure? Identify any exponential price
#surges and recommend the best booking window.
#b)Compare prices across airlines for the same route (e.g., Delhi-Mumbai). Which airlines are
#consistently cheaper/premium, and why?
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("flight_price.csv")

# a) Price vs Days Left
days_price = df.groupby("days_left")["price"].mean()
print("Average Price by Days Left:")
print(days_price)

# Plot
plt.figure(figsize=(8,4))
sns.lineplot(x=days_price.index, y=days_price.values)
plt.title("Flight Price vs Days Left")
plt.xlabel("Days Left")
plt.ylabel("Average Price")
plt.show()


# b) Airline Comparison (Delhi → Mumbai Example)
route_df = df[(df["source_city"]=="Delhi") & (df["destination_city"]=="Mumbai")]
airline_price = route_df.groupby("airline")["price"].mean().sort_values()
print("\nAverage Price by Airline (Delhi–Mumbai):")
print(airline_price)

# Plot
plt.figure(figsize=(8,4))
sns.barplot(x=airline_price.index, y=airline_price.values)
plt.title("Airline Price Comparison (Delhi–Mumbai)")
plt.xlabel("Airline")
plt.ylabel("Average Price")
plt.xticks(rotation=45)
plt.show()
```

(a) Days Left vs. Price

Prices increase as departure date approaches.

Last 7 days show exponential surge.

Book 30–45 days early for best price.

(b) Airline Price Comparison

Cheaper: IndiGo, Akasa Air → low-cost airlines.

Premium: Vistara, Air India → full-service airlines.

Prices differ due to service level, brand, and cost structure.

```python
#10:  HR Analytics Dataset
#a). What factors most strongly correlate with employee attrition? Use visualizations to show key
#drivers (e.g., satisfaction, overtime, salary).
#b). Are employees with more projects more likely to leave?

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("/content/hr_analytics.csv")


# a) Correlation with Attrition (left)
# Select only numeric columns for correlation
corr = df[['left','satisfaction_level','last_evaluation','number_project',
           'average_montly_hours','time_spend_company','Work_accident',
           'promotion_last_5years']].corr()

print("Correlation with Attrition (left):")
print(corr['left'].sort_values(ascending=False))

# Heatmap
plt.figure(figsize=(8,5))
sns.heatmap(corr, annot=True, cmap="Blues")
plt.title("Correlation Heatmap: Key Drivers of Attrition")
plt.show()


# b) Are employees with more projects leaving more?
project_attrition = df.groupby("number_project")["left"].mean()
print("\nAttrition Rate by Number of Projects:")
print(project_attrition)

# Plot
plt.figure(figsize=(6,4))
sns.lineplot(x=project_attrition.index, y=project_attrition.values)
plt.title("Attrition vs Number of Projects")
plt.xlabel("Number of Projects")
plt.ylabel("Attrition Rate")
plt.show()


```
